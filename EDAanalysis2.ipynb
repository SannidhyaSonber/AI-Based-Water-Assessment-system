```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the Excel file
file_path = '/content/newcombined4loc.xlsx'
df = pd.read_excel(file_path)

# Select numeric columns
numeric_cols = df.select_dtypes(include='number').columns.tolist()

# Check if there are any numeric columns
if not numeric_cols:
    print("No numeric columns found in the DataFrame.")
else:
    # Set up colors for each column
    colors = plt.cm.get_cmap('tab20', len(numeric_cols)).colors

    # Plot setup
    plt.figure(figsize=(20, 10))
    bar_width = 5 / len(numeric_cols)  # width of each bar

    # Plot bars with offsets
    for i, col in enumerate(numeric_cols):
        plt.bar(df.index + i * bar_width, df[col], width=bar_width, color=colors[i], label=col)

    plt.xlabel('Index')
    plt.ylabel('Values')
    plt.title('Combined Bar Graph of All Numeric Columns')
    plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1))
    plt.tight_layout()
    plt.show()
```
