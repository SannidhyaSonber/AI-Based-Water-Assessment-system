```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_excel('/content/newcombined4loc.xlsx')
df.head()
```

To fill the missing values in the dataset


```python


# List columns you want to exclude from filling
exclude_columns = ['date', '__Sheet__']  # Replace with your actual column names

# Select numeric columns except excluded ones
numeric_cols = [col for col in df.select_dtypes(include='number').columns if col not in exclude_columns]

# Fill missing values only in these numeric columns
for col in numeric_cols:
    mean = df[col].mean()
    std = df[col].std()
    nan_indices = df[col][df[col].isna()].index
    random_adjustment = np.random.uniform(-std, std, size=len(nan_indices))
    df.loc[nan_indices, col] = mean + random_adjustment

# Save the updated file if needed
df.to_excel('/content/newcombined4loc.xlsx', index=False)
```

```python
for col in df.columns:
    if col not in exclude_columns:
        # Try converting column to numeric, ignore non-convertible values
        df[col] = pd.to_numeric(df[col], errors='coerce')
        mean = df[col].mean()
        std = df[col].std()
        # Only fill if mean is not NaN
        if not np.isnan(mean):
            nan_indices = df[col][df[col].isna()].index
            random_adjustment = np.random.uniform(-std, std, size=len(nan_indices))
            df.loc[nan_indices, col] = mean + random_adjustment
```

```python
# Check for missing values in each column #Only for numerical values
missing_counts = df.isna().sum()

print("Missing values per column after filling:")
print(missing_counts)

# Optional: check if any missing values remain overall
if missing_counts.sum() == 0:
    print("No missing values remain in the dataset.")
else:
    print("Warning: Some missing values still remain.")
```
